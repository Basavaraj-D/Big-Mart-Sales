### Let us create an Artificial Neural Network for our Big Mart sales dataset and see how it will compare against the model we create in feature engineering and model building stage

#### 1.Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
import tensorflow as tf
```

- let us import the preprocessed file which we had earlier created.

```python
file = "/content/drive/MyDrive/Colab Notebooks/projects/Big mart sales/Big_mart_processed.csv"
data = pd.read_csv(file)
```

```python
data.head()
```

- let us drop all the features which will not be useful for the analysis

```python
df = data.copy()
```

```python
df.drop(['item_identifier','item_type','outlet_identifier','outlet_establishment_year','outlet_age_category'],axis=1,inplace=True)
```

```python
df.head()
```

#### 2.Train-Test-Split

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
print(X.shape,y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state =0)
```

    After Scaling the MSE increases so opted to not do scaling for this dataset

#### 3. Building the ANN

-  Initialising the ANN

```python
ann = tf.keras.models.Sequential()
```

- Adding the input layer and the firs hidden layer

```python
ann.add(tf.keras.layers.Dense(units=15, activation='relu'))
```

- Adding the second hidden layer

```python
ann.add(tf.keras.layers.Dense(units=15, activation='relu'))
```

- Adding the third hidden layer

```python
# ann.add(tf.keras.layers.Dense(units=7, activation='relu'))
```

- Adding the output layer

```python
ann.add(tf.keras.layers.Dense(units=1))
```

### 4.Training the ANN

```python
ann.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
ann.fit(X_train, y_train, batch_size = 32, epochs = 100)
```

```python
y_pred = ann.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import r2_score,mean_squared_error,accuracy_score
```

```python
r2_score(y_test,y_pred)
```

```python
MSE = mean_squared_error(y_test,y_pred)
print('we have an MSE of:',MSE)
```

```python
import math
RMSE = math.sqrt(MSE)
print('we have an RMSE of',RMSE)
```
